```python
from pathlib import Path
BASE_DIR = Path().resolve()
import sys
sys.path.append(str(BASE_DIR.parent / 'scr'))

from entrenamiento3 import entrenar_modelo

datasets = ['100000']

epochs = 1000
batch_size = 64
lr = 1e-4

for ds in datasets:

    X_path = BASE_DIR.parent / 'data' / f'dataset_{ds}_test_old' / 'X_train.npy'
    Y_path = BASE_DIR.parent / 'data' / f'dataset_{ds}_test_old' / 'Y_train.npy'

    X_val_path = BASE_DIR.parent / 'data' / f'dataset_{ds}_test_old' / 'X_val.npy'
    Y_val_path = BASE_DIR.parent / 'data' / f'dataset_{ds}_test_old' / 'Y_val.npy'

    subfolder_name = f'MLP_{ds}_test_old_run04'

    print(f'Entrenando en dataset {ds}...')
    entrenar_modelo(X_path, Y_path, X_val_path, Y_val_path, subfolder_name, epochs, batch_size, lr)
    print(f'Finalizado entrenamiento en dataset {ds}.\n')
```
