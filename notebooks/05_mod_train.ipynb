```python
from pathlib import Path
BASE_DIR = Path().resolve()
import sys
sys.path.append(str(BASE_DIR.parent / 'scr'))

from entrenamiento4 import entrenar_modelo

datasets = ['100000']

epochs = 1000
batch_size = 32
lr = 1e-4
radio = 1
alfa = 0.3

for ds in datasets:

    X_path = BASE_DIR.parent / 'data' / f'dataset_{ds}_test' / 'X_train.npy'
    Y_path = BASE_DIR.parent / 'data' / f'dataset_{ds}_test' / 'Y_train.npy'

    X_val_path = BASE_DIR.parent / 'data' / f'dataset_{ds}_test' / 'X_val.npy'
    Y_val_path = BASE_DIR.parent / 'data' / f'dataset_{ds}_test' / 'Y_val.npy'

    subfolder_name = f'MLP_{ds}_test_run12'

    print(f'Entrenando en dataset {ds}...')
    entrenar_modelo(X_path, Y_path, X_val_path, Y_val_path, subfolder_name, epochs, batch_size, lr, radio, alfa)
    #entrenar_modelo(X_path, Y_path, X_val_path, Y_val_path, subfolder_name, epochs, batch_size, lr)
    print(f'Finalizado entrenamiento en dataset {ds}.\n')
```
