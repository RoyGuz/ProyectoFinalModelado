```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

subfolder_name = "MLP_100000_test_run03"

BASE_DIR = Path().resolve()
save_folder = BASE_DIR.parent / 'results' / subfolder_name

loss_train = np.load(save_folder / 'loss_history_Train.npy')
loss_val = np.load(save_folder / 'loss_history_Val.npy')

plt.figure(figsize=(8, 5))
plt.loglog(loss_train, label='Train Loss', linewidth=2,linestyle = '--')
plt.loglog(loss_val, label='Validation Loss', linewidth=2, linestyle='--')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('Curva de Pérdida: Entrenamiento vs Validación')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
